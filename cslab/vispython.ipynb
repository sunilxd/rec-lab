```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv('company_sales_data.csv')
sales.head()
```

```python
x = sales['month_number']
plt.plot(x, sales['total_profit'], linestyle='dotted', label='Total profit')
plt.legend()
```

```python
products = sales.columns.values.tolist()[1:-2]

for prod in products:
    plt.plot(x, sales[prod], label=prod, marker='o')

plt.legend()
```

```python
plt.scatter(x, sales['toothpaste'], label='toothpaste')
plt.legend()
```

```python
plt.bar(x-.2, sales['facecream'], width=.3, label='facecream')
plt.bar(x+.2, sales['facewash'], width=.3, label='facewash')
plt.legend()
```

```python
plt.bar(x, sales['bathingsoap'], label='bathingsoap', width=.5)
plt.legend()
```

```python
plt.hist(sales['total_units'], label='total units')
plt.legend()
```

```python
products = sales.columns.values.tolist()[1:-2]

y = [sum(sales[prod]) for prod in products]
plt.pie(y, labels=products)
plt.show()
```
