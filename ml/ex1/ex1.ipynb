```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
```

```python
print('{}\n{}'.format(df.shape, df.head()))
```

```python
def plot_by(data_frame, column_name):
    uni_set = set(data_frame['variety'])
    dic_var = {ele:data_frame.loc[data_frame['variety']==ele] for ele in uni_set}
    for ele in dic_var:
        plt.scatter(dic_var[ele][column_name],np.zeros_like(dic_var[ele][column_name]))
    plt.xlabel(column_name)
    plt.show()
```

```python
for ele in list(df.head(0))[:-1]:
    plot_by(df, ele)
```

```python
sns.FacetGrid(df,hue='variety',height=5).map(plt.scatter,"sepal.width","petal.width").add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue='variety',height=5).map(plt.scatter,"sepal.length","petal.length").add_legend()
plt.show()
```
